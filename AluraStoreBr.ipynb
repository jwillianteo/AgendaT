### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

```python
# prompt: df.groupby('loja')['valor'].sum()

# Concatenate all DataFrames into one
df = pd.concat([loja, loja2, loja3, loja4])

# Group by 'loja' and sum the 'valor' column
df_sum = df.groupby('loja')['valor'].sum()
df_sum
```

```python
# prompt: import pandas as pd
# import matplotlib.pyplot as plt
# # Garantindo que todas as lojas tenham a coluna 'loja'
# loja['loja'] = 'Loja 1'
# loja2['loja'] = 'Loja 2'
# loja3['loja'] = 'Loja 3'
# loja4['loja'] = 'Loja 4'
# # Concatenando todas as lojas em um único DataFrame
# todas_lojas = pd.concat([loja, loja2, loja3, loja4])
# # Verificando se as colunas necessárias existem
# if 'produto' not in todas_lojas.columns or 'quantidade' not in todas_lojas.columns:
#     print("Erro: Verifique se as colunas 'produto' e 'quantidade' existem no DataFrame.")
# else:
#     # Agrupando por produto e somando a quantidade vendida
#     vendas_por_produto = todas_lojas.groupby('produto')['quantidade'].sum().sort_values(ascending=False)
#     # Produtos mais vendidos
#     produtos_mais_vendidos = vendas_por_produto.head(10).reset_index()
#     produtos_mais_vendidos.columns = ['Produto', 'Quantidade Vendida']
#     # Produtos menos vendidos
#     produtos_menos_vendidos = vendas_por_produto.tail(10).reset_index()
#     produtos_menos_vendidos.columns = ['Produto', 'Quantidade Vendida']
#     # Exibindo os DataFrames
#     print("Produtos Mais Vendidos:")
#     display(produtos_mais_vendidos)
#     print("\nProdutos Menos Vendidos:")
#     display(produtos_menos_vendidos)
#     # Gráfico dos mais vendidos
#     plt.figure(figsize=(10, 5))
#     plt.bar(produtos_mais_vendidos['Produto'], produtos_mais_vendidos['Quantidade Vendida'], color='green')
#     plt.title('Top 10 Produtos Mais Vendidos')
#     plt.xticks(rotation=45, ha='right')
#     plt.ylabel('Quantidade Vendida')
#     plt.tight_layout()
#     plt.show()
#     # Gráfico dos menos vendidos
#     plt.figure(figsize=(10, 5))
#     plt.bar(produtos_menos_vendidos['Produto'], produtos_menos_vendidos['Quantidade Vendida'], color='red')
#     plt.title('Top 10 Produtos Menos Vendidos')
#     plt.xticks(rotation=45, ha='right')
#     plt.ylabel('Quantidade Vendida')
#     plt.tight_layout()
#     plt.show()

# Assuming 'loja', 'loja2', 'loja3', and 'loja4' are dataframes with 'produto' and 'quantidade' columns.
# If they don't exist or have different column names, modify accordingly.


# Sample dataframes if the original ones are not available
data = {'produto': ['A', 'B', 'C', 'A', 'B'], 'quantidade': [10, 5, 15, 20, 8]}
loja = pd.DataFrame(data)
loja2 = pd.DataFrame(data)
loja3 = pd.DataFrame(data)
loja4 = pd.DataFrame(data)

# Garantindo que todas as lojas tenham a coluna 'loja'
loja['loja'] = 'Loja 1'
loja2['loja'] = 'Loja 2'
loja3['loja'] = 'Loja 3'
loja4['loja'] = 'Loja 4'
# Concatenando todas as lojas em um único DataFrame
todas_lojas = pd.concat([loja, loja2, loja3, loja4])
# Verificando se as colunas necessárias existem
if 'produto' not in todas_lojas.columns or 'quantidade' not in todas_lojas.columns:
    print("Erro: Verifique se as colunas 'produto' e 'quantidade' existem no DataFrame.")
else:
    # Agrupando por produto e somando a quantidade vendida
    vendas_por_produto = todas_lojas.groupby('produto')['quantidade'].sum().sort_values(ascending=False)
    # Produtos mais vendidos
    produtos_mais_vendidos = vendas_por_produto.head(10).reset_index()
    produtos_mais_vendidos.columns = ['Produto', 'Quantidade Vendida']
    # Produtos menos vendidos
    produtos_menos_vendidos = vendas_por_produto.tail(10).reset_index()
    produtos_menos_vendidos.columns = ['Produto', 'Quantidade Vendida']
    # Exibindo os DataFrames
    print("Produtos Mais Vendidos:")
    display(produtos_mais_vendidos)
    print("\nProdutos Menos Vendidos:")
    display(produtos_menos_vendidos)
    # Gráfico dos mais vendidos
    plt.figure(figsize=(10, 5))
    plt.bar(produtos_mais_vendidos['Produto'], produtos_mais_vendidos['Quantidade Vendida'], color='green')
    plt.title('Top 10 Produtos Mais Vendidos')
    plt.xticks(rotation=45, ha='right')
    plt.ylabel('Quantidade Vendida')
    plt.tight_layout()
    plt.show()
    # Gráfico dos menos vendidos
    plt.figure(figsize=(10, 5))
    plt.bar(produtos_menos_vendidos['Produto'], produtos_menos_vendidos['Quantidade Vendida'], color='red')
    plt.title('Top 10 Produtos Menos Vendidos')
    plt.xticks(rotation=45, ha='right')
    plt.ylabel('Quantidade Vendida')
    plt.tight_layout()
    plt.show()
```
